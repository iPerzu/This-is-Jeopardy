```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv('jeopardy.csv')
print(df.head)
```

```python
#rename columns
df.columns = ['Show Number', 'Air Date', 'Round', 'Category', 'Value',
       'Question', 'Answer']
df.columns
```

```python
#function filtering question with words(list)
def filter_question(data, words):
    filter = lambda x: all(word.lower() in x.lower() for word in words)
    return data.loc[data['Question'].apply(filter)]    

filter_question(df, ['King', "England"])
```

```python
#add Float Value column with floats from Value column
df["Float Value"] = df['Value'].apply(lambda x: float(x[1:].replace(',','')) if x != "None" else 0)
#print(df)
```
